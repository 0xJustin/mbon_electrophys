```python
import pandas as pd
mbon_skel = pd.read_csv('MBON-02-right.swc', comment='#', delim_whitespace=True, header=None, names=['n', 'type', 'x', 'y', 'z', 'radius', 'parent'])
```

```python
import networkx as nx
# Create a copy of mbon_skel and remove the root node
mbon_skel_edges = mbon_skel[mbon_skel['parent'] != -1].copy()

# Convert the parent column to integer type
mbon_skel_edges['parent'] = mbon_skel_edges['parent'].astype(int)

# Create a networkx graph from mbon_skel_edges
mbon_graph = nx.from_pandas_edgelist(mbon_skel_edges, 'n', 'parent', create_using=nx.DiGraph)
```

```python
import neuprint_functions
mbon_swc, relabel_dict = neuprint_functions.relabel_skeleton_swc(mbon_graph, mbon_skel, row_name="n", link_name="parent")
```

```python
import pandas as pd
from io import StringIO

# Convert the mbon_swc string into a pandas DataFrame
mbon_swc_df = pd.read_csv(StringIO(mbon_swc), comment='#', delim_whitespace=True, header=None, names=['n', 'type', 'x', 'y', 'z', 'radius', 'parent'])
```
